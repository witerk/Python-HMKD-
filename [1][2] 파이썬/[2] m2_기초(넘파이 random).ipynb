#### [Random 함수]  

- np.random.seed       # seed를 통한 난수 생성
- np.random.randint    # 균일분포의 정수 난수 1개 생성
- np.random.rand       # 0부터 1사이의 균일분포에서 난수 매트릭스 array 생성
- np.random.randn      # 가우시안 표준 정규 분포에서 난수 매트릭스 array 생성
- np.random.shuffle    # 기존의 데이터의 순서 바꾸기
- np.random.choice     # 기존의 데이터에서 sampling

```python
# 과제
# Q. 아래 가이드에 따라서 고객 구매데이터를 생성하고 데이터 프레임으로 변환하여 출력하세요.
# (관측치 1000개)

# id : 1번 ~ 1000번 일련번호
# gender : 0,1 정수 난수 생성
# age : 10 ~ 80사이 정수 난수를 생성
# region : 1 ~ 10 사이 정수 난수 생성
# prod_22 : 제품코드 11111 ~ 99999
# prod_23 : 제품코드 11111 ~ 99999
# price_22 : 1000 ~ 50000(원) 사이 난수 생성
# price_23 : 1000 ~ 50000(원) 사이 난수 생성
# qty_22 : 1 ~ 100 사이 정수 난수 생성하여('19년 구매 수량)
# qty_23 : 1 ~ 100 사이 정수 난수 생성하여('20년 구매 수량)
# time_22 : 01 ~ 24 사이 정수 난수 생성('19년에 가장 자주 구매한 시간대)
# time_23 : 01 ~ 24 사이 정수 난수 생성('20년에 가장 자주 구매한 시간대)
# amount_22: price_22 * qty_22
# amount_23: price_23 * qty_23
# sales : 22년 대비 23년 구매금액이 증가면 1, 감소면 0

npr.seed(5)
columns=['id','gender','age','region','prod_22','prod_23','price_22','price_23',
         'qty_22','qty_23','time_22','time_23','amount_22',
         'amount_23','sales']

df=pd.DataFrame(columns=columns)

for i in range(1000):
    id=i+1
    gender=npr.choice([0,1])
    age=npr.randint(10,80)
    region=npr.randint(1,11)
    prod_22=npr.randint(11111,100000,size=5)
    prod_23=npr.randint(11111,100000,size=5)
    price_22=npr.randint(1,51)*1000
    price_23=npr.randint(1,51)*1000
    qty_22=npr.randint(1,101)
    qty_23=npr.randint(1,101)
    time_22=npr.randint(1,25)
    time_23=npr.randint(1,25)
    amount_22=price_22 * qty_22
    amount_23=price_23 * qty_23
    sales=np.where(amount_22<amount_23, 1, 0)

    imsi=[id, gender, age, region, prod_22, prod_23, price_22, price_23, 
                        qty_22, qty_23, time_22, time_23, amount_22, amount_23, sales]
    
    df.loc[len(df)] = imsi
df.tail()
```

- 연도별로 상기한 사항들의 변동과 구매 감소간의 상관성을 부여하는 데이터 수정
- 데이터 수정은 불린 인덱싱으로 특정 그룹을 선택한 후 고객 특성에 맞는 값으로 변경

```python
import random
import numpy as np
from numpy import random as npr
import pandas as pd

npr.seed(10)
ar=np.zeros((1000,17),dtype=int)     #타입 안 정해주면 float 기본

ar[:,0]=np.arange(1,1001)     #id
ar[:,1]=np.random.randint(2,size=1000)        #gender
ar[:,2]=np.random.randint(1,91,size=1000)     #age
ar[:,3]=np.random.randint(1,11,size=1000)     #region
ar[:,4]=np.random.choice([11,22,33,44,55],1000)     #pdt_22 구매상품
ar[:,5]=np.random.choice([11,22,33,44,55],1000)     #pdt_23
ar[:,6]=np.random.randint(1,51,1000) * 1000     #pri_22 가격
ar[:,7]=np.random.randint(1,51,1000) * 1000     #pri_23
ar[:,8]=np.random.randint(1,101,1000)     #count_22 구매개수
ar[:,9]=np.random.randint(1,101,1000)     #count_23
ar[:,10]=np.random.randint(1,25,1000)     #time_22 가장 많이 구매한 시간대
ar[:,11]=np.random.randint(1,25,1000)     #time_23
ar[:,12]=np.random.randint(1,6,size=1000)     #channel 구매한 매장 유형
ar[:,13]=np.random.randint(1,11,size=1000)     #tr_year 구매이력 누적연수
ar[:,14]=ar[:,6]*ar[:,8]     #pur_22 구매총합
ar[:,15]=ar[:,7]*ar[:,9]     #pur_23
ar[:,16]=np.where(ar[:,15]>ar[:,14], 1, 0)     #sales (구매액 증가:1)

############## 데이터프레임 생성 ##############
col=['id','gender','age','region','pdt_22','pdt_23','pri_22','pri_23','ct_22','ct_23',
    'time_22','time_23','channel','tr_year','pur_22','pur_23','sales']

af=pd.DataFrame(ar, columns=col)
af=af.set_index('id')

af_org=af.copy()

############## NaN값 삽입 ##############
for i in range(len(af.iloc[0])-3):       #각 칼럼별 nan값 넣기(id, 구매액, 증감 제외)
    nan_c=random.sample(range(1000),10)
    for k in nan_c:
        af.iloc[k,i]=np.nan
```

```python
df=af

############## NaN값 처리 및 데이터프레임 수정 ##############
df.tr_year.replace(np.nan, 0, inplace=True) 
df.time_22.replace(np.nan, 0, inplace=True)
df.time_23.replace(np.nan, 0, inplace=True)
df.iloc[:,9:11]=df.iloc[:,9:11].astype('int')
df.iloc[:,12]=df.iloc[:,12].astype('int')
#구매 누적연도, 구매시간대 nan -> 0

df['pur_sum']=df['pur_22']+df['pur_23']
#22년, 23년 총합 매출액

df.dropna(subset=['pdt_22', 'pdt_23', 'ct_22', 'ct_23', 'pri_22', 'pri_23'], axis=0, inplace=True)
#구매상품, 개수, 가격 nan -> 삭제

df['ct_22']=df['ct_22'].astype('int')
df['ct_23']=df['ct_23'].astype('int')
#개수 int형으로 복구

df.age.replace(np.nan, -1, inplace=True)
df.age=df.iloc[:,1].astype('int')
#나이 nan -> -1로 변경 후 int형으로 복구

df['pdt_22'].replace({11:'크로커다일',22:'반스',33:'아디다스',44:'나이키',55:'퓨마'}, inplace=True)
df['pdt_23'].replace({11:'크로커다일',22:'반스',33:'아디다스',44:'나이키',55:'퓨마'}, inplace=True)
#구매상품 이름 변경

df.channel.replace({1:'브랜드매장',2:'백화점',3:'위탁',4:'팝업스토어',5:'인터넷',np.nan:'기타'}, inplace=True)
#매장 유형 nan -> '기타'로 변경

df.region.replace({1:'A지역',2:'B지역',3:'C지역',4:'D지역',5:'E지역',6:'F지역',7:'G지역',8:'H지역',9:'I지역',10:'J지역', np.nan:'그 외'}, inplace=True)
#지역 nan -> '기타'로 변경

df.gender.replace({0:'남성',1:'여성', np.nan:'정보없음'}, inplace=True)
#성별 nan -> 정보없음으로 변경

df.sales.replace({0:'감소',1:'증가'}, inplace=True)
#증감

############## 구매액 -> 등급 매기기(범주화) ##############
counts, bin_div = np.histogram(df.pur_22, bins=5)
counts2, bin_div2 = np.histogram(df.pur_23, bins=5)
names = ['bronze', 'silver', 'gold', 'vip', 'vvip']
df['level_22'] = pd.cut(x=df.pur_22, bins=bin_div, labels=names, include_lowest=True)
df['level_23'] = pd.cut(x=df.pur_23, bins=bin_div2, labels=names, include_lowest=True)

############## vvip회원들이 가장 많이 구매한 상품/ 선호하는 시간대 ##############
product22={'크로커다일':0,'반스':0,'아디다스':0,'나이키':0,'퓨마':0}
product23={'크로커다일':0,'반스':0,'아디다스':0,'나이키':0,'퓨마':0}
time22, time23=[0 for i in range(25)], [0 for i in range(25)]

for i in range(len(df)):
    if df.iloc[i, 17]=='vvip' or df.iloc[i,17]=='vip':
        product22[df.iloc[i,3]]+=1
        time22[df.iloc[i,9]]+=1

for i in range(len(df)):
    if df.iloc[i, 18]=='vvip' or df.iloc[i,18]=='vip':
        product23[df.iloc[i,4]]+=1
        time23[df.iloc[i,10]]+=1

pmax22=max(product22,key=product22.get)       #가장 많이 구매한 제품
pmin22=min(product22,key=product22.get)       #가장 적게 구매한 제품
pmax23=max(product23,key=product23.get)  
pmin23=min(product23,key=product23.get) 
print(f"22년도 VIP, VVIP 회원 최다-최소 구매 제품: {pmax22} / {pmin22}")
print(f"23년도 VIP, VVIP 회원 최다-최소 구매 제품: {pmax23} / {pmin23}\n")

best_t1 = list(filter(lambda x: time22[x] == max(time22), range(len(time22))))     #선호하는 시간(인덱스) 뽑기
best_t2 = list(filter(lambda x: time23[x] == max(time23), range(len(time23))))
#index는 동일한 수가 있을 경우 1개만 뽑아줘서 filter 사용

print(f"22년도 VIP, VVIP 회원들이 가장 선호하는 시간대: {best_t1}시")
print(f"23년도 VIP, VVIP 회원들이 가장 선호하는 시간대: {best_t2}시 \n")

############## 총 매출액이 가장 높은 지역 ##############
rg={'A지역':0,'B지역':0,'C지역':0,'D지역':0,'E지역':0,'F지역':0,'G지역':0,'H지역':0,'I지역':0,'J지역':0, '그 외':0}
rg_count={'A지역':0,'B지역':0,'C지역':0,'D지역':0,'E지역':0,'F지역':0,'G지역':0,'H지역':0,'I지역':0,'J지역':0, '그 외':0}
for i in range(len(df)):
    rg[df.iloc[i,2]]+=(df.iloc[i,13] + df.iloc[i,14])
    rg_count[df.iloc[i,2]]+=1

del rg['그 외']
del rg_count['그 외']

for i in rg:
    rg[i]=round(rg[i]/rg_count[i],2)     #각 지역별 인당 평균 매출액

pmax=max(rg,key=rg.get)
pmin=min(rg,key=rg.get)

print(f"1인당 총 구매액이 가장 많은 지역: {pmax} \n1인당 총 구매액이 가장 적은 지역: {pmin}\n")
# print(rg)

############## 고객 누적 연도/시간대에 따른 총 매출액 ##############  의미없는 분석 그래프
# df[['pur_sum','tr_year']].plot(kind='scatter', x='tr_year',y='pur_sum')
# df[['pur_sum','time_22']].plot(kind='scatter', x='time_22',y='pur_sum')

############## 감소 고객 중 성별 ############## 
gen={'여성':0,'남성':0, '정보없음':0}

for i in range(len(df)):
    if df.iloc[i, 15]=='감소':
        gen[df.iloc[i,0]]+=1
        
gmax=max(gen,key=gen.get)
print(f"22년 대비 23년 {gmax} 고객의 구매 감소 비율이 더 높습니다.\n")


col=['gender','age','region','pdt_22','pdt_23','pri_22','pri_23','ct_22','ct_23',
    'time_22','time_23','channel','tr_year','pur_22','level_22','pur_23','level_23','pur_sum','sales']
df=df[col]

df.isnull().sum()
# df.info()
df
# df.corr()
```
