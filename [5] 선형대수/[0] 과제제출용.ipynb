#### 과제(1): 5차원 배열 2개를 생성하여 행렬곱을 계산하세요.

```python
a = np.random.rand(2,1,4,2,3)
b = np.random.rand(2,3,4,3,2)
c = np.matmul(a,b)
print(f"행렬곱한 결과(형상): {c.shape}\n")
print(c)

# 5차원 행렬의 경우 결과값의 형상이 1~3차원까지는 같고,
# (이때 두 배열의 1~3차원의 크기가 같지 않으므로 브로드캐스팅 됨,
#  더 높은 차원 쪽(2,3,4)을 따름)
# 4차원은 a의 4번째 차원(2), 5차원은 b의 마지막 차원(2) 값이 나옴
# => (2,1,4,2,3)*(2,3,4,3,2) = (2,3,4,2,2)

# 4차원 배열도 마찬가지
# 맨 뒤 2개 행렬에 따라 결과값의 형상이 정해짐
# 1~2차원은 동일하게, 3차원은 a의 3차원 크기, 4차원은 b의 마지막 차원 크기
```

#### 과제(2): 아래 그래프의 x,y구간을 400으로 조밀하게 변경하여 다시 표현

```python
def f(x):
    return x**3 - 3*x**2 + x

x= np.linspace(-1,3,400)
y=f(x)

plt.figure(figsize=(10,6))
plt.plot(x,y, 'b-')

plt.xlim(-2,4)
plt.xticks(np.arange(-1,4))
plt.yticks(np.arange(-5,4))
plt.title('함수: $f(x) = x^3 - 3x^2 + x$의 그래프')

plt.show()
```

#### 과제(3): 다음 함수의 역함수를 구하고 그래프로 표현하세요.
- 원래 함수의 x와 y를 바꿔줍니다.
- 바뀐 함수를 y에 대해서 푼 후 x로 정리합니다.

```python
def f(x):
    return (x**2+1)
def f_inv(x):
    return np.sqrt(x-1)

x=np.linspace(0,10,300)
plt.figure(figsize=(10,6))
plt.plot(x, f(x), 'r-', label='함수 $f(x) = x^2+1$')
plt.plot(x, f_inv(x), 'b-', label='역함수 $f^{-1}(x) = \sqrt{x-1}$')
plt.plot(x,x,'g--')

plt.axis('equal')
plt.xlim(1,3)
plt.ylim(0,3)
plt.legend()
plt.title('역함수 그래프')
plt.show()
```

#### 과제(4): 로그함수를 시각화하여 표현하세요

```python
def logx(x):
    return np.log(x)

x = np.linspace(0, 6, 100)
y = logx(x)

plt.figure(figsize=(10,6))
plt.plot(x,y, color='skyblue')

plt.xlim(-1,6)
plt.ylim(-4,5)
plt.title("$logX$ 그래프")
plt.show()
```
