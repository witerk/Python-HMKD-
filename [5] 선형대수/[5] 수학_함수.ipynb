#### 역함수
- 어던 함수의 입력/출력 관계와 정반대의 입출력 관계를 갖는 함수
- 원래의 함수에서 x축과 y축이 바뀐 것이므로 y=x가 나타내는 직선을 대칭축으로 대칭인 함수의 그래프

```python
def f1(x):
    return x**2
def f1inv(x):
    return np.sqrt(x)

x=np.linspace(0,3,300)
plt.figure(figsize=(10,6))
plt.plot(x, f1(x), 'r-', label='함수 $f(x) = x^2$')
plt.plot(x, f1inv(x), 'b-', label='역함수 $f^{-1}(x) = \sqrt{x}$')
plt.plot(x,x,'g--')

plt.axis('equal')
plt.xlim(0,2)
plt.ylim(0,2)
plt.legend()
plt.title('역함수 그래프')
plt.show()
```

```python
# ReLU 함수(활성화 함수)
# : 입력값에 대해 0 이하는 0, 0보다 큰 값은 그대로 출력
xx = np.linspace(-10,10,100)
plt.figure(figsize=(10,6))
plt.plot(xx, np.maximum(xx,0))

plt.title('max(x,0) 또는 ReLU')
plt.xlabel('$x$')
plt.ylabel('$ReLU$')
plt.show()
```

#### 과제(3): 다음 함수의 역함수를 구하고 그래프로 표현하세요.
- 원래 함수의 x와 y를 바꿔줍니다.
- 바뀐 함수를 y에 대해서 푼 후 x로 정리합니다.

```python
def f(x):
    return (x**2+1)
def f_inv(x):
    return np.sqrt(x-1)

x=np.linspace(0,10,300)
plt.figure(figsize=(10,6))
plt.plot(x, f(x), 'r-', label='함수 $f(x) = x^2+1$')
plt.plot(x, f_inv(x), 'b-', label='역함수 $f^{-1}(x) = \sqrt{x-1}$')
plt.plot(x,x,'g--')

plt.axis('equal')
plt.xlim(1,3)
plt.ylim(0,3)
plt.legend()
plt.title('역함수 그래프')
plt.show()
```
